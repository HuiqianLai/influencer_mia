```python
import json
import csv

# 读取JSON文件
with open('/Users/huiqianlai/Documents/GitHub/influencer_mia/youtube_downloads_test/transcripts_all.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

# 合并字幕，按视频维度
merged_transcripts = []

for video in data:
    video_id = video['video_id']

    # 获取所有字幕文本
    all_text = []
    for transcript in video['transcripts']:
        language = transcript['language']
        for segment in transcript['segments']:
            text = segment['text'].strip()
            if text:  # 只添加非空文本
                all_text.append(text)

    # 合并成一个完整文本
    merged_text = ' '.join(all_text)

    # 去掉所有换行符和回车符
    merged_text = merged_text.replace('\n', ' ').replace('\r', ' ')

    # 去掉多余的空格（可选）
    merged_text = ' '.join(merged_text.split())

    merged_transcripts.append({
        'video_id': video_id,
        'full_transcript': merged_text
    })

# 保存为CSV文件
with open('merged_transcripts.csv', 'w', encoding='utf-8', newline='') as f:
    writer = csv.DictWriter(f, fieldnames=['video_id', 'full_transcript'])
    writer.writeheader()  # 写入表头
    writer.writerows(merged_transcripts)

# 打印结果预览
print(f"已成功保存到 merged_transcripts.csv")
print(f"总共处理了 {len(merged_transcripts)} 个视频\n")

for item in merged_transcripts:
    print(f"Video ID: {item['video_id']}")
    print(f"Transcript length: {len(item['full_transcript'])} characters")
    print(f"Preview: {item['full_transcript'][:100]}...")
    print("-" * 80)
```
